```python
import pandas as pd
import numpy as np
```

## Read datasource for H-1B salaries


```python
fb_h1b = pd.read_csv('../Data/Facebook_Salaries_2018.csv')
ggle_h1b = pd.read_csv('../Data/Google_Salaries_2018.csv')
sf_h1b = pd.read_csv('../Data/SalesForce_Salaries_2018.csv')
wkd_h1b = pd.read_csv('../Data/Workday_Salaries_2018.csv')

```

```python
fb_h1b.head()
#fb_h1b.dtypes
#ggle_h1b.head()
#sf_h1b.head()
#wkd_h1b.head()
```

```python
#add column for place of work
fb_h1b['Place_of_Work'] = 'Facebook' 
ggle_h1b['Place_of_Work'] = 'Google' 
sf_h1b['Place_of_Work'] = 'SalesForce'
wkd_h1b['Place_of_Work'] = 'Workday' 
```

## Group data by job role and find the average pay (salary) for each

```python
#group by job role
fb_grouped = fb_h1b.groupby(['Job_Role'])

#find average salary by job role
fb_salary_avg = fb_grouped['Pay'].mean()

## this didn't work, just have dataframe with salary
#fb_PlaceofWork = fb_h1b['Place_of_Work']
#fb_HireDate = fb_h1b['Hire_Date']
#fb_Year = fb_h1b['Hire_Date_Yr']
#fb_JobTitle = fb_h1b['Job_Role']

```

```python
#let's look at it
fb_salary_avg
```

## Make to Data Frame

```python
#now make dataframe out of grouped information
fb_h1b_grouped = pd.DataFrame({
                               'Salary': fb_salary_avg})
```

```python
fb_h1b_grouped
```

## Now merge Salary Data Frame to the original fb_h1b Data Frame 

```python
fb_h1b_grouped_df = pd.merge(fb_h1b_grouped, fb_h1b, on = 'Job_Role', how ='left')

fb_h1b_grouped_df
```
